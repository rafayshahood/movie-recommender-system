```python
import pandas as pd
import numpy as np

movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

### Choosing columns to be kept


1.   genres
2.   id
3.   keywords
4.   title
5.   overview
6.   cast
7.   crew




```python
movies = movies[['id', 'title', 'genres', 'keywords', 'overview', 'cast', 'crew']]
movies.head()
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.iloc[0].genres
```

```python
movies.iloc[0].cast
```

```python
def convert3(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter < 3:
      l.append(i['name'])
      counter +=1
    else:
      break
  return l
```

```python
 movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.iloc[0].cast
```

```python
movies['crew'][0]
```

```python
def getDirectorName(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(getDirectorName)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# converting like this:
# 'Sam Worthington'
# 'SamWorthington'
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(' ', '') for i in x])
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head(2)
```

```python
newMovies = movies[['id', 'title', 'tags']]
newMovies.head()
```

```python
newMovies['tags'] = newMovies['tags'].apply(lambda x: " ".join(x))
```

```python
newMovies['tags'][0]
```

```python
newMovies['tags'] = newMovies['tags'].apply(lambda x: x.lower())
```

### Vectorization

```python
# stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
newMovies['tags'] = newMovies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(newMovies['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarityScores = cosine_similarity(vectors)
```

```python
similarityScores[0]
```

```python
def recommend(movie):
  movieIndex = newMovies[newMovies['title'] == movie ].index[0]
  similarity = similarityScores[movieIndex]
  moviesList = sorted(list(enumerate(similarity)), reverse = True, key= lambda x: x[1])[1:6]

  for i in moviesList:
    print(newMovies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
